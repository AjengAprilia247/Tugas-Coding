```python
import numpy as np
```

```python
# buat vector i dengan elemen (1,0) dan vector j dengan elemen (0,1)
i = np.array([1, 0])
print(i)
```

```python
j = np.array([0, 1])
print(j)
```

```python
 # bahwa vector i dan vector j itu orthogonal
 np.dot(i, j)
```

```python
# Bagaimana dengan v=(1,2) dan w=(2,-1), apakah juga orthogonal?
v = np.array([1, 2])
w = np.array([2, -1])
print(v)
print(w)
```

```python
# buktikan bahwa vector v dan vector w itu orthogonal
np.dot(v, w)
```
